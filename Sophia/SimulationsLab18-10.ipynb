```python
# programm to simulate computer infection

import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# 0 = Healthy 
# 1 = Infected

# variables
N = 1000
system = []
healthyNodes = 0 
infectedNodes = 0
currentInfected = 0 

infectedArray1 = []
```

```python
system = np.zeros((1000), dtype=int)

x = random.randint(1, N)
# print(x)
system[x] = 1
currentInfected = currentInfected + 1
infectedNodes = infectedNodes + 1
infectedArray1.append(currentInfected)

# print(system)
```

```python
# print(currentInfected)
while infectedNodes < N:
  for i in range (0,currentInfected):
    infect = random.randint(0, N-1)
    if system[infect] == 0:
      system[infect] = 1
      currentInfected += 1
      infectedNodes += 1 
  infectedArray1.append(currentInfected)
```

```python
# print(system)
```

```python
rateGrouth1 = []
for i in range (len(infectedArray1)-1):
  rate = infectedArray1[i+1] - infectedArray1[i]
  rateGrouth1.append(rate)

print(rateGrouth1)
plt.plot(rateGrouth1)
plt.title("Infection Rate per Step")
plt.show()

print(infectedArray1)
plt.plot(infectedArray1)
plt.title("Infected per Step")
plt.show()
```
