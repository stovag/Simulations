```python
# earthquakes
import random


fout = 0.001
fcrit = 4
```

```python
rows, cols = (9, 9)
first_arr = [[0 for i in range(cols)] for j in range(rows)]
second_arr = [[0 for i in range(cols)] for j in range(rows)]
# 3rd_arr = [[0]*cols]*rows
```

```python
for i in range(rows):
  for j in range(cols):

    randnum = random.randint(0,3999)/1000
    first_arr[i][j] = randnum
    first_arr[0][0] = 3.999

print(first_arr)

```

```python
for i in range(5):
  for i in range(rows):
    for j in range(cols):
      if first_arr[i][j] < fcrit:
        first_arr[i][j] = first_arr[i][j] + fout
      else:
        # να φτιάξω τα edges
        if (i+1>rows) and (j+1>cols):
          first_arr[i-1][j] = first_arr[i-1][j] + 1
          first_arr[i][j-1] = first_arr[i][j-1] + 1
          first_arr[i][j] = 0
          second_arr[i][j] = 1
        elif i+1>10:
          first_arr[i][j+1] = first_arr[i][j+1] + 1
          first_arr[i][j+1] = first_arr[i][j+1] + 1
          first_arr[i-1][j] = first_arr[i-1][j] + 1
          first_arr[i][j] = 0
          second_arr[i][j] = 1
        elif j+1>10:
          first_arr[i+1][j] = first_arr[i+1][j] + 1
          first_arr[i-1][j] = first_arr[i-1][j] + 1
          first_arr[i][j-1] = first_arr[i][j-1] + 1
          first_arr[i][j] = 0
          second_arr[i][j] = 1
        else:
          first_arr[i][j+1] = first_arr[i][j+1] + 1
          first_arr[i][j+1] = first_arr[i][j+1] + 1
          first_arr[i-1][j] = first_arr[i-1][j] + 1
          first_arr[i+1][j] = first_arr[i+1][j] + 1
          first_arr[i][j] = 0
          second_arr[i][j] = 1



print(second_arr)
print(first_arr)
```
