```python
# import libraries
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
size_of_array_x = 20
size_of_array_y = 20
full_size = 20*20
# informed_nodes = 0
current_informed = 0
initial_array = [[0 for i in range(size_of_array_x)] for j in range(size_of_array_y)]

informed_nodes = []
```

```python
x = random.randint(0,19)
y = random.randint(0,19)
print(x,y)
initial_array[x][y] = 1
# informed_nodes += 1
current_informed += 1

print(initial_array)
```

```python
print(full_size)
while current_informed < full_size:
  x = random.randint(0,19)
  y = random.randint(0,19)

  if initial_array[x][y] == 0:
    initial_array[x][y] = 1
    current_informed += 1
  informed_nodes.append(current_informed)
  print(len(informed_nodes))
```

```python
plt.plot(informed_nodes)
plt.title("Infected per Step")
plt.show()
```
