```python
import numpy as np
import random as r
from matplotlib import pyplot as plt

b = 1
succ_chance = 10
n = 10000
nodes = np.zeros(n)
nodes[r.randint(0,n-1)] = 1
infected = 1
infections = []

while infected < len(nodes):
    for i in range(infected):
        for j in range(b):
            target = r.randint(0, n-1)
            die = r.randint(1, 100)
            if nodes[target] == 0 and die <= succ_chance:
                nodes[target] = 1
                infected += 1
    infections.append(infected)
    
d = [infections[i]-infections[i-1] for i in range(1, len(infections))]
print("Steps:", len(infections))    
    
plt.plot(infections)
plt.title("Infected per Step")
plt.show()

plt.plot(d)
plt.title("Infection Rate per Step")
plt.show()
```
